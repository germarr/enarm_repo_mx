```python
import pandas as pd
import duckdb
import os
from requests_html import HTMLSession
from openai import OpenAI
import json
import re

from dotenv import load_dotenv
load_dotenv()
```

```python
years = ['2024','2023']

export_df = []

for year in years:
    hospital_csv = pd.read_csv(f'/Users/gerardomartinez/Desktop/Projects/enarm/additional_docs/centros_salud/{year}.csv')
    tamagochi = hospital_csv.to_dict('records')
    final_dataframe = []

    for i in tamagochi:
        # Extract hospital name
        hospital = i['Centro']

        # Create DataFrame
        df = pd.DataFrame([
            {'Centro': hospital, 'Especialidad': key, 'Valor': value}
            for key, value in i.items()
            if key not in ['Centro', 'TOTAL']
        ])
        final_dataframe.append(df)

    hospitals_df = pd.concat(final_dataframe).reset_index(drop=True)
    hospitals_df['year'] = year

    export_df.append(hospitals_df)

# Concatenate all DataFrames
hospitals_dataframe = pd.concat(export_df).reset_index(drop=True)
```
